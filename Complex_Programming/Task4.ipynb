```python
import matplotlib.pyplot as plt

plt.rcParams["font.size"] = 20
plt.rcParams['figure.figsize'] = 12,10 
 
# Growth Function
# N for Pray and P for Predator
def do_logistic_growth(NA_0, NB_0, a, s, r_A, d_A, r_B, d_B, P_0, T):
    #Initialize    
    NA_t = NA_0
    NB_t = NB_0
    X = NA_t+NB_t
    P_t = P_0
    i = 0
    
    #P for Predator, A for Prey A, B for Prey B N for Prey and T for Time
    dataP = []
    dataA = []
    dataB = []
    dataN = []
    dataT = []
    
    dataP.append(P_t)
    dataA.append(NA_t)
    dataB.append(NB_t)
    dataN.append(X)
    dataT.append(i)
    
    for i in range(T-1):
        Q = d_A * NA_t + d_B * NB_t
        NA_tt = NA_t + a * NA_t * (r_A - X - d_A * P_t)
        NB_tt = NB_t + a * NB_t * (r_B - X - d_B * P_t)
        P_tt = P_t + a * P_t * (Q -s)
        X_t = NA_tt + NB_tt
        
        dataP.append(P_tt)
        dataA.append(NA_tt)
        dataB.append(NB_tt)
        dataN.append(X_t)
        dataT.append(i+1)
        
        NA_t = NA_tt
        NB_t = NB_tt
        P_t = P_tt
        X = X_t
        
    return(dataP, dataN, dataA, dataB, dataT)


# Set parameter
a = 0.2
s = 0.1
NA_0 = 0.1
NB_0 = 0.1
P_0 = 0.1
T = 3000

r_A = 0.9
r_B = 0.8
d_A = 0.5
d_B = 0.1

dataP, dataN, dataA, dataB, dataT = do_logistic_growth(NA_0, NB_0, a, s, r_A, d_A, r_B, d_B, P_0, T)

plt.subplot(4,1,1)
plt.xlabel("Time")
plt.ylabel("Predato")
plt.plot(dataT, dataP)
plt.legend(['Predator'])

plt.subplot(4,1,2)
plt.xlabel("Time")
plt.ylabel("Prey A")
plt.plot(dataT, dataA)
plt.legend(['Prey A'])

plt.subplot(4,1,3)
plt.xlabel("Time")
plt.ylabel("Prey B")
plt.plot(dataT, dataB)
plt.legend(['Prey B'])

plt.subplot(4,1,4)
plt.xlabel("Time")
plt.ylabel("Prey")
plt.plot(dataT, dataN)
plt.legend(['Prey'])

plt.show()
```
