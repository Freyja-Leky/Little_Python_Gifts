Followed Code is for 課題１：典型的挙動の可視化

Here A for a:単調増加->収束, B for b:振動->減衰, C for c:不規則

```python
import matplotlib.pyplot as plt

plt.rcParams["font.size"] = 20
plt.rcParams['figure.figsize'] = 12,10 
 
# Growth Function 
def do_logistic_growth(k, r, n0, t):
    nt= n0
    datax=[]
    datay=[]
 
    datax.append(0)
    datay.append(nt)
 
    for i in range(t):
        nt= nt+r*nt*(1.0-nt/k)
        datax.append(i+1)
        datay.append(nt)
 
    return(datax, datay)

# Set K，N_0, T
K = 100.0
N_0 = 1.0
T = 20

#Test from 0.0 to 1.0
# for i in range (0,10,1):
#     j = i/10 
#     dataX, dataY= do_logistic_growth(K,j,N_0,T)
#     s = 'r = '+ str(j)
#     plt.plot(dataX, dataY,label = s)

#Test from 1.0 to 2.0
# for i in range (10,20,1):
#     j = i/10 
#     dataX, dataY= do_logistic_growth(K,j,N_0,T)
#     s = 'r = '+ str(j)
#     plt.plot(dataX, dataY,label = s)

#Test from 2.0 to 3.0
# for i in range (20,30,1):
#     j = i/10 
#     dataX, dataY= do_logistic_growth(K,j,N_0,T)
#     s = 'r = '+ str(j)
#     plt.plot(dataX, dataY,label = s)

#A : r = 0.8
dataX, dataY= do_logistic_growth(K,0.8,N_0,T)
plt.plot(dataX, dataY,label='A : r = 0.8')

#B : r = 1.8
dataX, dataY= do_logistic_growth(K,1.8,N_0,T)
plt.plot(dataX, dataY,label='B : r = 1.8')    

#C : r = 2.8
dataX, dataY= do_logistic_growth(K,2.8,N_0,T)
plt.plot(dataX, dataY,label='C : r = 2.8')

#Display the Graph
plt.legend()    
plt.xlabel("time")
plt.ylabel("population size")
plt.grid()
plt.show()
```

Followed Code is for 課題２：分岐図を描く

```python
import matplotlib.pyplot as plt

plt.rcParams["font.size"] = 20
plt.rcParams['figure.figsize'] = 12,10 
 
# Growth Function 
# Here record r and N_t if t>250
def do_logistic_growth(k, r, n0, t):
    nt= n0
    datax=[]
    datay=[] 
    for i in range(t):
        nt= nt+r*nt*(1.0-nt/k)
        if i > 250:
            datax.append(r)
            datay.append(nt)
 
    return(datax, datay)

# Set K，N_0, T
K = 100.0
N_0 = 1.0
T = 300

for i in range(100,301,1):
    i_c = i/100
    dataX, dataY= do_logistic_growth(K,i_c,N_0,T)
    plt.plot(dataX, dataY,'.')
     
plt.xlabel("r")
plt.ylabel("N_t")
plt.grid()
plt.show()
```
