### Conclusion about Prey and Predator Code
Here the graph of prey and predator shows that the quantity change of both is stable in a loop. As the population of prey decreases, at first predator will increase. However, when the prey population reaches one level, the amount of predator will also be reduced. Same as the predator to the pray. And finally they will change together in a larger loop with more or less population.

```python
import matplotlib.pyplot as plt

plt.rcParams["font.size"] = 20
plt.rcParams['figure.figsize'] = 12,10 
 
# Growth Function
# N for Prey and P for Predator
def do_logistic_growth(a, r, s, n0, p0, t):
    nt = n0
    pt = p0
    
    datax=[]
    datay=[]
    dataT=[]
 
    datax.append(nt)
    datay.append(pt)
    dataT.append(0)
 
    for i in range(t):
        ntt = nt+a*nt*(r-pt)
        ptt = pt+a*pt*(nt-s)
        if ntt > 0 and ntt < 1 and ptt > 0 and ptt < 1:
            datax.append(ntt)
            datay.append(ptt)
            dataT.append(i)
        pt = ptt
        nt = ntt
 
    return(datax, datay, dataT)

# Set parameter
a = 0.2
N_0 = 0.11
P_0 = 0.1
T = 1000

dataX, dataY, dataT= do_logistic_growth(a, 0.1, 0.1, N_0, P_0, T)


#Display the Graph  
plt.subplot(2,1,2)
plt.xlabel("prey size")
plt.ylabel("predator size")
plt.plot(dataX, dataY)

plt.subplot(2,1,1)
plt.plot(dataT, dataX, label='prey')
plt.plot(dataT, dataY, label='predator')
plt.xlabel("Time")
plt.ylabel("Size")
plt.legend() 

plt.show()



```
