```python
import pandas as pd
from openpyxl import Workbook
from openpyxl.styles import Font, colors, Alignment, Border, Side

#sort data
score = pd.read_excel('./Score.xlsx',header=0,index_col=0)
score['Row_sum'] = score.sum(axis=1)
score.sort_values(by='Row_sum',ascending=False,inplace=True)
#score.drop('Row_sum')
score.loc['Col_sum'] = score.sum()
score.sort_values('Col_sum',axis=1,ascending=False,inplace=True)
rn=score.pop('Row_sum')
score.insert(score.shape[1],'Row_sum',rn)

#write to excel
writer = pd.ExcelWriter('sp.xlsx',engine = 'openpyxl')
score.to_excel(writer,sheet_name="SP")
worksheet = writer.sheets['SP']

#create border type
#only s type
s_blue = '0000FF'
p_red = 'FF0000'
sp_purple = '8A2BE2'

#only s type
bordertype1 = Border(right=Side(border_style='thick', color=s_blue))
bordertype2 = Border(right=Side(border_style='thick', color=s_blue),
                    bottom=Side(border_style='thick', color=s_blue))
bordertype3 = Border(bottom=Side(border_style='thick', color=s_blue))

#only p type
bordertype4 = Border(right=Side(border_style='thick', color=p_red))
bordertype5 = Border(right=Side(border_style='thick', color=p_red),
                    bottom=Side(border_style='thick', color=p_red))
bordertype6 = Border(bottom=Side(border_style='thick', color=p_red))

#s1+p
bordertype7 = Border(right=Side(border_style='thick', color=sp_purple))
bordertype8 = Border(right=Side(border_style='thick', color=sp_purple),
                    bottom=Side(border_style='thick', color=p_red))
bordertype9 = Border(right=Side(border_style='thick', color=s_blue),
                    bottom=Side(border_style='thick', color=p_red))

#s2+p
bordertype10 = Border(right=Side(border_style='thick', color=sp_purple),
                    bottom=Side(border_style='thick', color=s_blue))
bordertype11 = Border(right=Side(border_style='thick', color=sp_purple),
                    bottom=Side(border_style='thick', color=sp_purple))
bordertype12 = Border(right=Side(border_style='thick', color=s_blue),
                    bottom=Side(border_style='thick', color=sp_purple))

#s3+p
bordertype13 = Border(right=Side(border_style='thick', color=p_red),
                    bottom=Side(border_style='thick', color=s_blue))
bordertype14 = Border(right=Side(border_style='thick', color=p_red),
                    bottom=Side(border_style='thick', color=sp_purple))
bordertype15 = Border(bottom=Side(border_style='thick', color=sp_purple))


#merge s&p
def merge_type(s,p):
    if s == 1:
        if p == 4:
            return 7
        elif p == 5:
            return 8
        elif p == 6:
            return 9
    elif s == 2:
        if p == 4:
            return 10
        elif p == 5:
            return 11
        elif p == 6:
            return 12
    elif s == 3:
        if p == 4:
            return 13
        elif p == 5:
            return 14
        elif p == 6:
            return 15


        
ExcelType = score.copy(deep = True)
for i in range(0,ExcelType.shape[0]-1):
    for j in range(0,ExcelType.shape[1]-1):
        ExcelType.iloc[i,j]=0
#print(ExcelType)

#decide s
for r in range(0,score.shape[0]-1):
    rs = score.iloc[r,score.shape[1]-1]
    if (rs !=0):
        ExcelType.iloc[r,rs-1]=1
#for r in range(0,score.shape[0]-1):
    if (score.iloc[r,score.shape[1]-1] > score.iloc[r+1,score.shape[1]-1]):
        down = score.iloc[r+1,score.shape[1]-1]
        top = score.iloc[r,score.shape[1]-1]
        for l in range(down,top):
            if (ExcelType.iloc[r,l] == 1):
                ExcelType.iloc[r,l]=2
            else:
                ExcelType.iloc[r,l]=3
    elif(score.iloc[r,score.shape[1]-1] < score.iloc[r+1,score.shape[1]-1]):
        down = 0
        top = score.iloc[r,score.shape[1]-1]
        for l in range(down,top):
            if (ExcelType.iloc[r,l] == 1):
                ExcelType.iloc[r,l]=2
            else:
                ExcelType.iloc[r,l]=3
#print(ExcelType)

ExcelP = score.copy(deep = True)
for i in range(0,ExcelP.shape[0]-1):
    for j in range(0,ExcelP.shape[1]-1):
        ExcelP.iloc[i,j]=0
#print(ExcelP)

#decide p
for l in range(0,score.shape[1]-1):
    cs = score.iloc[score.shape[0]-1,l]
    if (cs !=0):
        ExcelP.iloc[cs-1,l]=6
    if (score.iloc[score.shape[0]-1,l] > score.iloc[score.shape[0]-1,l+1]):
        down = score.iloc[score.shape[0]-1,l+1]
        top = score.iloc[score.shape[0]-1,l]
        for r in range(down,top):
            if (ExcelP.iloc[r,l] == 6):
                ExcelP.iloc[r,l]= 5
            else:
                ExcelP.iloc[r,l]= 4
    elif (score.iloc[score.shape[0]-1,l] < score.iloc[score.shape[0]-1,l+1]):
        down = 0
        top = score.iloc[score.shape[0]-1,l]
        for r in range(down,top):
            if (ExcelP.iloc[r,l] == 6):
                ExcelP.iloc[r,l]= 5
            else:
                ExcelP.iloc[r,l]= 4
#print(ExcelP)

for i in range(0,ExcelType.shape[0]-1):
    for j in range(0,ExcelType.shape[1]-1):
        if (ExcelType.iloc[i,j] != 0 and ExcelP.iloc[i,j]!= 0):
            print(ExcelType.iloc[i,j])
            print(ExcelP.iloc[i,j])
            ExcelType.iloc[i,j]=merge_type(ExcelType.iloc[i,j],ExcelP.iloc[i,j])
        elif (ExcelType.iloc[i,j] == 0 and ExcelP.iloc[i,j]!= 0):
            ExcelType.iloc[i,j]=ExcelP.iloc[i,j]
print(ExcelType)

def loadtype(et):
    if (et == 1):
        return bordertype1
    elif (et == 2):
        return bordertype2
    elif (et == 3):
        return bordertype3
    elif (et == 4):
        return bordertype4
    elif (et == 5):
        return bordertype5
    elif (et == 6):
        return bordertype6
    elif (et == 7):
        return bordertype7
    elif (et == 8):
        return bordertype8
    elif (et == 9):
        return bordertype9
    elif (et == 10):
        return bordertype10
    elif (et == 11):
        return bordertype11
    elif (et == 12):
        return bordertype12
    elif (et == 13):
        return bordertype13
    elif (et == 14):
        return bordertype14
    elif (et == 15):
        return bordertype15
    else:
        return
    
#score['Row_percent'] = score.apply(lambda x:x.Row_sum/(ExcelType.shape[1]-1),axis = 1)     
    
for i in range(1,score.shape[0]):
    for j in range(1,score.shape[1]):
        worksheet.cell(row=i+1,column=j+1).border = loadtype(ExcelType.iloc[i-1,j-1])

     
        
writer.save()

print(score)
```
