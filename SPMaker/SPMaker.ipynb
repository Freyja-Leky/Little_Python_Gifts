3. Run this program
4. New file named "SP.xlsx" will be created as the SP table of the data table
* SP table will be created with sum, weight and percentage

```python
#read data
score = pd.read_excel('./Score.xlsx',header=0,index_col=0)
data_row = score.shape[0]
data_col = score.shape[1]
```

#### Calculate&Sort part
1. Sum (Descending)
2. Weight (Sum up the score or amount of the student which is 0 / Ascending) 
3. Percentage 

```python
#sum
score['Row_sum'] = score.sum(axis=1)
score.loc['Col_sum'] = score.sum()

#weight
def row_weight(x):
    if (x.name == 'Col_sum' or x.name == 'Col_weight'):
        return ""
    ans = 0
    for i in range (0,len(x)):
        if (x.iat[i] == 0):
            ans = ans + score.loc['Col_sum'].iat[i]
    return ans

score['Row_weight'] = score.apply(lambda x:row_weight(x),axis = 1)

def col_weight(x):
    if (x.name == 'Row_sum' or x.name == 'Row_weight'):
        return ""
    ans = 0
    for i in range (0,len(x)):
        if (x.iat[i] == 0):
            ans = ans + score.iloc[i].at['Row_sum']
    return ans

score.loc['Col_weight'] = score.apply(lambda x:col_weight(x),axis = 0)

#sort
score.sort_values(['Row_sum','Row_weight'],ascending = [False,True],inplace = True)
score.sort_values(['Col_sum','Col_weight'],axis = 1,ascending = [False,True],inplace = True)

#arrange the table - sum
score.drop('Col_sum',inplace = True)
score.drop('Row_sum',axis = 1,inplace = True)
score.drop('Col_weight',inplace = True)
score.drop('Row_weight',axis = 1,inplace = True)
score['Row_sum'] = score.sum(axis=1)
score.loc['Col_sum'] = score.sum()

#copy for style
ExcelType = score.copy(deep = True)
for i in range(0,data_row):
    for j in range(0,data_col):
        ExcelType.iloc[i,j]=0
ExcelP = ExcelType.copy(deep = True)

#arrange the table - weight
score['Row_weight'] = score.apply(lambda x:row_weight(x),axis = 1)
score.loc['Col_weight'] = score.apply(lambda x:col_weight(x),axis = 0)

#percentage
def col_percentage(x):
    if (x.name == 'Row_percentage'):
        return ('%.2f%%')%((score.at['Col_sum','Row_sum']/(data_col*data_row))*100)
    elif(x.name == 'Row_sum' or x.name == 'Row_weight'):
        return ""
    else:
        return ('%.2f%%')%((x['Col_sum']/data_row)*100) 

score.loc['Col_percentage'] = score.apply(lambda x:col_percentage(x),axis=0)    

def row_percentage(x):
    if (x.name == 'Col_percentage'):
        return ('%.2f%%')%((score.at['Col_sum','Row_sum']/(data_col*data_row))*100)
    elif(x.name == 'Col_sum' or x.name == 'Col_weight'):
        return ""
    else:
        return ('%.2f%%')%((x['Row_sum']/data_col)*100)   

score['Row_percentage'] = score.apply(lambda x:row_percentage(x),axis=1)

print (score)
```

#### Style part
1.BorderType is the line of SP table, with S line in blue and P line in red, mixed line in purple  
2.Fill is the background color of labels, with student label in shallow blue and problem label in shallow red, calculation label in shallow purple


```python
#color
sl_blue = '0000FF'
pl_red = 'FF0000'
spl_purple = '8A2BE2'
border_line = 'thick'

sb_blue = 'F0F8FF'
pb_red = 'FFE4E1'
spb_purple = 'E6E6FA'
background_type = 'solid'

#set line type
#only s type
bordertype1 = Border(right=Side(border_style=border_line, color=sl_blue))
bordertype2 = Border(right=Side(border_style=border_line, color=sl_blue),
                    bottom=Side(border_style=border_line, color=sl_blue))
bordertype3 = Border(bottom=Side(border_style=border_line, color=sl_blue))
#only p type
bordertype4 = Border(right=Side(border_style=border_line, color=pl_red))
bordertype5 = Border(right=Side(border_style=border_line, color=pl_red),
                    bottom=Side(border_style=border_line, color=pl_red))
bordertype6 = Border(bottom=Side(border_style=border_line, color=pl_red))
#s1+p
bordertype7 = Border(right=Side(border_style=border_line, color=spl_purple))
bordertype8 = Border(right=Side(border_style=border_line, color=spl_purple),
                    bottom=Side(border_style=border_line, color=pl_red))
bordertype9 = Border(right=Side(border_style=border_line, color=sl_blue),
                    bottom=Side(border_style=border_line, color=pl_red))
#s2+p
bordertype10 = Border(right=Side(border_style=border_line, color=spl_purple),
                    bottom=Side(border_style=border_line, color=sl_blue))
bordertype11 = Border(right=Side(border_style=border_line, color=spl_purple),
                    bottom=Side(border_style=border_line, color=spl_purple))
bordertype12 = Border(right=Side(border_style=border_line, color=sl_blue),
                    bottom=Side(border_style=border_line, color=spl_purple))
#s3+p
bordertype13 = Border(right=Side(border_style=border_line, color=pl_red),
                    bottom=Side(border_style=border_line, color=sl_blue))
bordertype14 = Border(right=Side(border_style=border_line, color=pl_red),
                    bottom=Side(border_style=border_line, color=spl_purple))
bordertype15 = Border(bottom=Side(border_style=border_line, color=spl_purple))

#merge s&p
def merge_type(s,p):
    if s == 1:
        if p == 4:
            return 7
        elif p == 5:
            return 8
        elif p == 6:
            return 9
    elif s == 2:
        if p == 4:
            return 10
        elif p == 5:
            return 11
        elif p == 6:
            return 12
    elif s == 3:
        if p == 4:
            return 13
        elif p == 5:
            return 14
        elif p == 6:
            return 15

#set label background
s_fill = PatternFill(fill_type = background_type, fgColor = sb_blue)
p_fill = PatternFill(fill_type = background_type, fgColor = pb_red)
c_fill = PatternFill(fill_type = background_type, fgColor = spb_purple)
```

#### Draw SP line part

```python
#decide s
for r in range(0,data_row):
    rs = ExcelType.iloc[r,data_col]
    if (rs != 0):
        ExcelType.iloc[r,rs-1] = 1
    if (ExcelType.iloc[r,data_col] > ExcelType.iloc[r+1,data_col]):
        down = ExcelType.iloc[r+1,data_col]
        top = ExcelType.iloc[r,data_col]
        for l in range(down,top):
            if (ExcelType.iloc[r,l] == 1):
                ExcelType.iloc[r,l] = 2
            else:
                ExcelType.iloc[r,l] = 3
    elif(ExcelType.iloc[r,data_col] < ExcelType.iloc[r+1,data_col]):
        down = 0
        top = ExcelType.iloc[r,data_col]
        for l in range(down,top):
            if (ExcelType.iloc[r,l] == 1):
                ExcelType.iloc[r,l] = 2
            else:
                ExcelType.iloc[r,l] = 3
#decide p
for l in range(0,data_col):
    cs = score.iloc[data_row,l]
    if (cs !=0):
        ExcelP.iloc[cs-1,l]=6
    if (ExcelP.iloc[data_row,l] > ExcelP.iloc[data_row,l+1]):
        down = ExcelP.iloc[data_row,l+1]
        top = ExcelP.iloc[data_row,l]
        for r in range(down,top):
            if (ExcelP.iloc[r,l] == 6):
                ExcelP.iloc[r,l] = 5
            else:
                ExcelP.iloc[r,l] = 4
    elif (ExcelP.iloc[data_row,l] < ExcelP.iloc[data_row,l+1]):
        down = 0
        top = ExcelP.iloc[data_row,l]
        for r in range(down,top):
            if (ExcelP.iloc[r,l] == 6):
                ExcelP.iloc[r,l] = 5
            else:
                ExcelP.iloc[r,l] = 4
#merage type
for i in range(0,data_row):
    for j in range(0,data_col):
        if (ExcelType.iloc[i,j] != 0 and ExcelP.iloc[i,j]!= 0):
            ExcelType.iloc[i,j]=merge_type(ExcelType.iloc[i,j],ExcelP.iloc[i,j])
        elif (ExcelType.iloc[i,j] == 0 and ExcelP.iloc[i,j]!= 0):
            ExcelType.iloc[i,j]=ExcelP.iloc[i,j]

print(ExcelType)
```

#### Output part

```python
#write to SP.xlsx
writer = pd.ExcelWriter('SP.xlsx',engine = 'openpyxl')
score.to_excel(writer,sheet_name="SP Table")
worksheet = writer.sheets['SP Table']

#load style
def loadtype(et):
    if (et == 1):
        return bordertype1
    elif (et == 2):
        return bordertype2
    elif (et == 3):
        return bordertype3
    elif (et == 4):
        return bordertype4
    elif (et == 5):
        return bordertype5
    elif (et == 6):
        return bordertype6
    elif (et == 7):
        return bordertype7
    elif (et == 8):
        return bordertype8
    elif (et == 9):
        return bordertype9
    elif (et == 10):
        return bordertype10
    elif (et == 11):
        return bordertype11
    elif (et == 12):
        return bordertype12
    elif (et == 13):
        return bordertype13
    elif (et == 14):
        return bordertype14
    elif (et == 15):
        return bordertype15
    else:
        return
for i in range(1,data_row+1):
    for j in range(1,data_col+1):
        worksheet.cell(row=i+1,column=j+1).border = loadtype(ExcelType.iloc[i-1,j-1])
        
calculation = 3
for i in range (2,data_row+2):
    worksheet.cell(row = i,column = 1).fill = s_fill
for j in range (1,data_col+1):
    worksheet.cell(row = 1,column = j+1).fill = p_fill
for i in range (data_row+2,data_row+2+calculation):
    worksheet.cell(row = i,column = 1).fill = c_fill
for j in range (data_col+1,data_col+1+calculation):
    worksheet.cell(row = 1,column = j+1).fill = c_fill

#output    
writer.save()
```
